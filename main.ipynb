By extracting movies from the two most similar users, the code recommends 5 movies that will user most likely find interesting. First, we made a function that returns all users similar to the user given as an input: 

```python
#the function that extracts all similar users of the user
def get_matching_users(user_id):
    
    # extract minh_results of the user
    user_data = df[df['user_id'] == user_id].drop_duplicates(subset=['minh_results'])
    minh_results = user_data['minh_results'].tolist()
    
    # extract all movie indices that have at least one similar bucket
    matching_movies_indices = df.apply(lambda row: any(all(row['minh_results'][i] == minh_result[i]
                                                           for i in range(len(minh_result)))
                                                       for minh_result in minh_results), axis=1)
    
    # extract all movies based on the indices
    matching_movies = df[matching_movies_indices]
    
    # excluding the movies from the own user
    matching_movies = matching_movies[matching_movies['user_id'] != user_id]
    
    # add column with total movies count for every user as a tie breaker
    matching_movies['movies_count'] = matching_movies.groupby('user_id')['user_id'].transform('count')
    
    # create temporary column of minh_results as a string for dropping duplicates
    matching_movies['tmp_minh_results'] = matching_movies['minh_results'].apply(lambda x: str(x))
    
    # extract unique minh_results from every user
    matching_users = matching_movies.drop_duplicates(subset=['user_id', 'tmp_minh_results'])
    
    # drop temporary column again
    unique_matching_users = matching_users.drop('tmp_minh_results', axis=1)
    
    return unique_matching_users
```

After finding two most similar users, the code extracts all the movies both of them watched. If there are 5 movies like that, these 5 movies will be recommended. If there are less then 5, other movies recommended will be taken on account of the number of clicks the most similar user made. If there are more then 5 movies two most similar users have in common, the 5 out of them will be chosen on account of the number of clicks of the most similar user.

```python
from collections import Counter
def get_movies(user_id):
    # Filter the DataFrame to include only rows where 'user_id' is the input user_id
    user_movies = df[df['user_id'] == user_id]

    # Get the 'movie_id' column from the filtered DataFrame
    movie_ids = user_movies['movie_id']

    return movie_ids.tolist()
#function that takes a user and returns five movies recommended 
def rec_movies(user_id):
    # extract all similar users of the user
    unique_matching_users = get_matching_users(user_id)

    # group by user_id and movies_count
    unique_matching_users = unique_matching_users.groupby(['user_id', 'movies_count']).agg({'minh_results': 'count'})

    # sort by the most unique similar minh_results and use movies_count as tie breaker
    similar_users = unique_matching_users.reset_index().sort_values(by=['minh_results', 'movies_count'], ascending=[False, False])
    su=list(similar_users['user_id'][0:2])


    m1=get_movies(su[0])
    m2=get_movies(su[1])
    movies=list(set(m1).intersection(set(m2)))
    if(len(movies)==5):
        return movies
    #we have to add movies from the most similar user
    if(len(movies)<5):
        for i in m1:
            if(i in movies):
                m1.remove(i) 
        c=Counter(m1)
        a=len(movies) 
        for i in c.most_common(5-a):
            movies.append(i[0])
        return movies
    #we have to leave top 5 of the most similar user among these movies he has in common with the second most similar
    if(len(movies)>5):
        for i in m1:
            if(i not in movies):
                m1.remove(i)
        c=Counter(m1)
        movies=list()
        for i in c.most_common(5):
            movies.append(i[0])
        return movies
```

```python
findspark.init()
spark = SparkSession.builder.master("local[4]").getOrCreate()

famd_dataset.to_pickle('dataset_2_famd')
dataset_point_2_famd = pd.read_pickle("dataset_2_famd")

schema = StructType([StructField("C1", DoubleType(), True), StructField("C2", DoubleType(), True),
                    StructField("C3", DoubleType(), True), StructField("C4", DoubleType(), True),
                    StructField("C5", DoubleType(), True), StructField("C6", DoubleType(), True),
                    StructField("C7", DoubleType(), True), StructField("C8", DoubleType(), True), StructField("C9", DoubleType(), True)])

sdf = spark.createDataFrame(dataset_point_2_famd, schema=schema)

k = 5
rdd = sdf.rdd.map(list)
vectorized_rdd = rdd.map(lambda row: np.array(row))

def calculate_distance(x, y):
    distances = np.linalg.norm(x - y, axis=1)
    key = np.argmin(distances)
    return key, x

def kmeans_iteration(centroids, data_rdd, k):
    distances_rdd = data_rdd.map(lambda row: calculate_distance(row, centroids))
    aggregated_data = distances_rdd.combineByKey(
        lambda x: (x, 1),
        lambda acc, x: (acc[0] + x, acc[1] + 1),
        lambda acc1, acc2: (acc1[0] + acc2[0], acc1[1] + acc2[1]))
    updated_centroids = aggregated_data.mapValues(lambda acc: acc[0] / acc[1]).collect()
    K_new = {i[0]: i[1] for i in updated_centroids}
    finalist = [np.array(K_new.get(i, None)) for i in range(k) if K_new.get(i, None) is not None]
    final_centroids = np.array(finalist)
    return final_centroids

initial_centroids = np.array(dataset_point_2_famd.sample(k))
difference = 999

while difference > 0.1:
    updated_centroids = kmeans_iteration(initial_centroids, vectorized_rdd, k)
    difference = np.linalg.norm(updated_centroids - initial_centroids)
    initial_centroids = updated_centroids

print(initial_centroids)
```

This Python script utilizes PySpark to implement the K-means clustering algorithm on the FAMD dataset with 9 columns, obtained before. The code initializes a Spark session, prepares the dataset, and converts it into a PySpark DataFrame. Just after this, the spark dataframe is converted to RDD and then to a vectorized RDD.
The function 'calculate_distance' is created to compute the Euclidean distance between 2 points in a vector space, this function is used to find the closest centroid for each data point. After this, the function kmeas_iteration represents the iteration of the k-means clustering algorithm, using the calculate_distance function, calculating the distances to all centroids. The data points are then aggregated and then new centroid positions are calculated by dividing the sum of data points for each centroid by their count. The while loop continues iterating if the variable 'difference' is greater than 0.1, in other terms, the k-means runs until the centroids converge. Then the final centroids are printed.

#### Method 1: Elbow

```python
import matplotlib.pyplot as plt

def calculate_inertia(centroids, data_rdd):
    distances_rdd = data_rdd.map(lambda row: calculate_distance(row, centroids))
    inertia = distances_rdd.map(lambda x: np.linalg.norm(x[1] - centroids[x[0]]))
    return inertia.sum()

inertias = []

for k_value in range(2, 10):
    initial_centroids = np.array(dataset_point_2_famd.sample(k_value))
    difference = 999

    while difference > 0.1:
        updated_centroids = kmeans_iteration(initial_centroids, vectorized_rdd, k_value)
        difference = np.linalg.norm(updated_centroids - initial_centroids)
        initial_centroids = updated_centroids

    inertia = calculate_inertia(initial_centroids, vectorized_rdd)
    inertias.append(inertia)
    print(f"K = {k_value}, Inertia = {inertia}")

plt.plot(range(2, 10), inertias, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

The first technique used to find the optimal number of clusters is the 'Elbow method', the main idea behind it, is to run the k-means for a range of values of k and then polt the sum of squared ditsances or variance for each k. The elbow in the final plot represents a a point where increasing the number of clusters does not has a significat impact on the reduction of the SSD, meaning that that number of cluster could be an optimal quantity. Implementging this method, the results is that 5 is a good number of clsuters.

The Silhouette method is the second techinque that has been chosen to find the optimal number of clusters, it evaluates the quontity of clusters by measuring how similar an object is to its own clsuter, compared to other clsuters, a highier value of the silhouette score indicates a better-defined clsuter. It varies from -1 to 1, where a score near 1 indiates that the object is well matched to its own cluster and not well matched to nieghbouring clusters while a silhouette score near 0 indicates overlapping clusters, and negative values suggest that data points might be assigned to the wrong cluster.
```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
import numpy as np
import random

def closest_centroid(point, centroids):
    distances = [np.linalg.norm(point - centroid) for centroid in centroids]
    return np.argmin(distances)

def kmeans(rdd, centroids, iterations=10):
    for _ in range(iterations):
        assigned_centroids = rdd.map(lambda point: (closest_centroid(point, centroids), (point, 1)))
        centroid_sum_count = assigned_centroids.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
        new_centroids = centroid_sum_count.mapValues(lambda x: x[0] / x[1]).collect()
        for idx, new_centroid in new_centroids:
            centroids[idx] = new_centroid
    return centroids

def kmeans_plusplus_init(rdd, k):
    centroids = [rdd.takeSample(False, 1, seed=random.randint(0, 1000))[0]]
    for _ in range(1, k):
        distances_rdd = rdd.map(lambda point: min(np.linalg.norm(point - c) ** 2 for c in centroids))
        total_distance = distances_rdd.sum()
        probabilities = distances_rdd.map(lambda distance: distance / total_distance).collect()
        new_centroid_index = np.random.choice(len(probabilities), p=probabilities)
        new_centroid = rdd.zipWithIndex().filter(lambda x: x[1] == new_centroid_index).map(lambda x: x[0]).first()
        centroids.append(new_centroid)
    return centroids


sc = SparkContext.getOrCreate()
spark = SparkSession(sc)

schema = StructType([StructField("C1", DoubleType(), True), StructField("C2", DoubleType(), True),
                    StructField("C3", DoubleType(), True), StructField("C4", DoubleType(), True),
                    StructField("C5", DoubleType(), True), StructField("C6", DoubleType(), True),
                    StructField("C7", DoubleType(), True), StructField("C8", DoubleType(), True), StructField("C9", DoubleType(), True)])
df = spark.createDataFrame(dataset_point_2_famd, schema=schema)

features = df.select("C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9")
data_rdd = features.rdd.map(lambda row: np.array([row[0], row[1]]))

k = 5

initial_centroids = kmeans_plusplus_init(data_rdd, k)

final_centroids = kmeans(data_rdd, initial_centroids)

print("Final centroids:")
for centroid in final_centroids:
    print(centroid)
```

This code implements a k-means algorithm using again PySpark with a k-means ++ inizialization. First, thanks to the function 'closest_centroid' the index of the closest centroid for a given data point is found, then data points are iteratively assigned to the closest centroids by the function 'kmeans' and the initial centroids are selected using a weighted probability distribution. After creating a SparkContext and a SparkSession the number of clusters are set at 5 and the k-means algorithm is run with the initialsed centroids to obtain the final cluster centroids.

## Recursive Algorithm

```python
# a) Recursive Algorithm

def find_max_score(S, exams):
    """
    Recursively finds the maximum score Federico can achieve.

    :param S: Federico's current score.
    :param exams: A list of exam scores.
    :return: The maximum score that can be achieved.
    """
    # Base case: If there are no exams left, return the current score
    if not exams:
        return S

    best_score = 0
    # Iterate through each exam
    for i in range(len(exams)):
        exam_score = exams[i]
        remaining_exams = exams[:i] + exams[i+1:]

        if exam_score < S:
            # If the exam is easy, increase the scores of remaining exams
            adjusted_exams = [score + (S - exam_score) for score in remaining_exams]
        else:
            # If the exam is hard, decrease the scores of remaining exams
            adjusted_exams = [score - (exam_score - S) for score in remaining_exams]

        # Recursively find the best score after taking this exam
        score_after_exam = find_max_score(exam_score, adjusted_exams)
        best_score = max(best_score, score_after_exam)

    return best_score

# Example usage
print("Maximum possible score for each scenario:")
print(find_max_score(8, [5, 7, 1]))
print(find_max_score(25, [18, 24, 21, 32, 27]))
print(find_max_score(30, [13, 27, 41, 59, 28, 33, 39, 19, 52, 48, 55, 79]))
```

This part implements a recursive algorithm to solve the exam score maximization problem. The function find_max_score takes Federico's current score and a list of exam scores as input and recursively calculates the maximum score he can achieve.

### Algorithm:

1. Base Case: If there are no exams left, return the current score.
   
2. Recursive Step: Iterate over each exam. For each exam, calculate the new score and adjust the scores of the remaining exams based on whether the exam is easier or harder than the current score. Then, recursively call the function with the new score and the adjusted list of remaining exams.
   
3. Maximization: After each recursive call, compare the returned score with the current maximum and update it if necessary.
   
4. Return: The function returns the maximum score achievable.

## b) Detailed Time Complexity Analysis for Recursive Solution


The time complexity of the recursive algorithm find_max_score can be analyzed as follows:

Function Calls: The function makes recursive calls by iterating over the list of exams, which has a length of N. In each call, it iterates through all the remaining exams, leading to N iterations in the first call, N-1 in the second, and so on.

Depth of Recursion: The depth of recursion corresponds to the number of exams. In the worst case, the function will recurse until it has processed all exams, leading to a maximum depth of N.

Work Done at Each Call: At each recursive call, the function performs operations like slicing lists and iterating over them. These operations have a time complexity of O(N) in the worst case.

Total Complexity: Combining the number of calls and the work done in each, the total time complexity becomes O(N * N!), as for each of the N exams, it explores all permutations of the remaining N-1 exams.

This factorial time complexity indicates that the algorithm is highly inefficient for large input sizes, as the execution time grows extremely fast with the increase in the number of exams.

## c) Optimized Algorithm with Dynamic Programming


```python
# c) Optimized Algorithm with Dynamic Programming

def find_max_score_dp(S, exams, memo):
    """
    Uses dynamic programming to find the maximum score, optimizing the recursive approach.

    :param S: Federico's current score.
    :param exams: A list of exam scores.
    :param memo: A dictionary used for memoization to store intermediate results.
    :return: The maximum score that can be achieved.
    """
    # Using tuple for exams as keys in memo, since lists are not hashable
    exams_key = tuple(exams)
    if (S, exams_key) in memo:
        return memo[(S, exams_key)]

    # Base case: No exams left
    if not exams:
        return S

    max_possible_score = 0
    # Iterate through each exam
    for i, p in enumerate(exams):
        new_exams = exams[:i] + exams[i+1:]
        if p < S:
            adjusted_exams = [x + (S - p) for x in new_exams]
        else:
            adjusted_exams = [x - (p - S) for x in new_exams]

        # Store the best score in memo and update the maximum score
        max_possible_score = max(max_possible_score, find_max_score_dp(p, adjusted_exams, memo))

    memo[(S, exams_key)] = max_possible_score
    return max_possible_score

def find_max_score_wrapper(S, exams):
    """
    Wrapper function for the dynamic programming solution.

    :param S: Initial score of Federico.
    :param exams: List of marks for all exams.
    :return: The maximum possible score.
    """
    memo = {}
    return find_max_score_dp(S, exams, memo)

# Example usage
print("Maximum possible score for each scenario with DP:")
print(find_max_score_wrapper(8, [5, 7, 1]))
print(find_max_score_wrapper(25, [18, 24, 21, 32, 27]))
print(find_max_score_wrapper(30, [13, 27, 41, 59, 28, 33, 39, 19, 52, 48, 55, 79]))
```

This part improves upon the recursive solution by introducing dynamic programming. The function find_max_score_dp employs memoization to store and reuse the results of subproblems, thereby significantly reducing the number of redundant computations.

Algorithm:

1. Memoization: Use a dictionary to store the results of subproblems.
    
2. Recursive Calls with Caching: Similar to the recursive approach, but check the memoization dictionary before computing a subproblem. If the result is already computed, return it from the dictionary.

3. Maximization and Memoization: After each recursive call, update the memoization dictionary with the new maximum score for the current state.
                                                                                                                     
4. Return: The function returns the maximum score achievable, using memoization to avoid redundant calculations.

## AI Model-Optimized Implementation

```python
# d) AI Model-Optimized Implementation

def find_max_score_optimized(S, exams):
    """
    Optimized algorithm to find the maximum possible score, using an AI model-generated approach.

    :param S: Initial score of Federico.
    :param exams: List of marks for all exams.
    :return: The maximum possible score.
    """
    memo = {}

    def dfs(S, exams):
        if not exams:
            return S

        exams_tuple = tuple(sorted(exams))
        if (S, exams_tuple) in memo:
            return memo[(S, exams_tuple)]

        max_score = 0
        for i, p in enumerate(exams):
            new_exams = exams[:i] + exams[i+1:]
            if p < S:
                adjusted_exams = [x + (S - p) for x in new_exams]
            else:
                adjusted_exams = [x - (p - S) for x in new_exams]

            max_score = max(max_score, dfs(p, adjusted_exams))

        memo[(S, exams_tuple)] = max_score
        return max_score

    return dfs(S, exams)

# Example usage
print("Maximum possible score for each scenario with optimized algorithm:")
print(find_max_score_optimized(8, [5, 7, 1]))
print(find_max_score_optimized(25, [18, 24, 21, 32, 27]))
print(find_max_score_optimized(30, [13, 27, 41, 59, 28, 33, 39, 19, 52, 48, 55, 79]))
```

This part presents an AI model-generated optimized solution. The algorithm uses a depth-first search (DFS) approach with improved memoization, aiming to efficiently navigate through the possible choices of exams.

Algorithm:

1. Improved Memoization: Use a dictionary to cache results, keying on the current score and a sorted tuple of remaining exams to recognize equivalent states.

2. Depth-First Search (DFS): Explore each possible path of taking exams, adjusting the scores of the remaining exams accordingly.

3. Maximization and Caching: Update the maximum score after each path exploration and store it in the memoization dictionary.

4. Return: The function returns the maximum score achievable, optimized through DFS and enhanced memoization.
